```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib

from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
useBackslash = False
trainingDatasetPath = r'..\data\training_dataset.csv' if useBackslash else r'../data/training_dataset.csv'
dataset_training = pd.read_csv(trainingDatasetPath, sep=';')
```

```python
print(dataset_training.columns.tolist())
```

```python
useBackslash = False
testDatasetPath = r'..\data\test_dataset.csv' if useBackslash else r'../data/test_dataset.csv'
df_test = pd.read_csv(testDatasetPath, sep=";")
```

```python
y_train = dataset_training['movie_score'].values
dataset_training = dataset_training.drop(columns=['movie_score', 'Unnamed: 0', 'averageRating', 'numVotes', '_orig_order'])
x_train = dataset_training.values
```

```python
y_test = df_test['movie_score'].values
df_test = df_test.drop(columns=['movie_score', 'Unnamed: 0', 'averageRating', 'numVotes', '_orig_order'])
x_test = df_test.values
```

```python
# Creating and training a simple linear regression model
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
# print scores
print("Score on training set: {:.3f}".format(lr.score(x_train, y_train)))
print("Score on test set: {:.3f}".format(lr.score(x_test, y_test)))
```

# Lasso regression with GridSearch over alpha (no polynomial features)

```python
lasso = Lasso(max_iter=1000)
param_grid_lasso = {
    'alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10]
}
lasso_grid_search = GridSearchCV(lasso, param_grid_lasso, cv=10, scoring='r2', n_jobs=-1)
lasso_grid_search.fit(x_train, y_train)
```

```python
# Inspect best Lasso alpha and scores
best_alpha_lasso = lasso_grid_search.best_params_['alpha']
best_lasso = lasso_grid_search.best_estimator_

print(f"Best Lasso alpha: {best_alpha_lasso}")
print(f"Best Lasso CV R-squared: {lasso_grid_search.best_score_:.3f}")
print(f"Best Lasso training R-squared: {best_lasso.score(x_train, y_train):.3f}")
print(f"Best Lasso test R-squared: {best_lasso.score(x_test, y_test):.3f}")
```

```python
# Ridge regression with GridSearch over alpha (no polynomial features)
ridge = Ridge(max_iter=10000)
param_grid_ridge = {
    'alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10]
}
ridge_grid_search = GridSearchCV(ridge, param_grid_ridge, cv=10, scoring='r2', n_jobs=-1)
ridge_grid_search.fit(x_train, y_train)
```

```python
# Inspect best Ridge alpha and scores
best_alpha_ridge = ridge_grid_search.best_params_['alpha']
best_ridge = ridge_grid_search.best_estimator_

print(f"Best Ridge alpha: {best_alpha_ridge}")
print(f"Best Ridge CV R-squared: {ridge_grid_search.best_score_:.3f}")
print(f"Best Ridge training R-squared: {best_ridge.score(x_train, y_train):.3f}")
print(f"Best Ridge test R-squared: {best_ridge.score(x_test, y_test):.3f}")
```

```python
# Creating and training a polynomial regression model (fixed degree, no GridSearch)
poly_degree = 3  # choose the polynomial degree manually
poly_lr = Pipeline([
    ("poly_features", PolynomialFeatures(degree=poly_degree)),
    ("linear_regression", LinearRegression()),
])

# Fit the model directly
poly_lr.fit(x_train, y_train)

# Evaluate on train and test
print(f"Polynomial degree: {poly_degree}")
print("Poly LR training R-squared: {:.3f}".format(poly_lr.score(x_train, y_train)))
print("Poly LR test R-squared: {:.3f}".format(poly_lr.score(x_test, y_test)))
```

```python
# Creating and training a polynomial Lasso regression model (fixed degree, no GridSearch)
poly_lasso_degree = 3   # choose polynomial degree manually
lasso_alpha = 0.001     # choose regularization strength manually

poly_lasso = Pipeline([
    ("poly_features", PolynomialFeatures(degree=poly_lasso_degree)),
    ("lasso_regression", Lasso(alpha=lasso_alpha, max_iter=10000)),
])

# Fit the model directly
poly_lasso.fit(x_train, y_train)

# Evaluate on train and test
print(f"Polynomial Lasso degree: {poly_lasso_degree}, alpha: {lasso_alpha}")
print("Poly Lasso training R-squared: {:.3f}".format(poly_lasso.score(x_train, y_train)))
print("Poly Lasso test R-squared: {:.3f}".format(poly_lasso.score(x_test, y_test)))
```
